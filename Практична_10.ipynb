```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import least_squares
import math

def func(x):
    return math.e**(math.cos(x))

x = np.array([i * 0.1 for i in range(1, 11)])
y = np.array([func(xi) for xi in x])
print('x =', x)
print('y =', y)

# Наближення прямою
def fun(a, x, y):
    return a[0] + a[1] * x - y

a0 = np.array([1, 1])
res_lsq = least_squares(fun, x0=a0, args=(x, y))
print("a0 = %.2f, a1 = %.2f" % tuple(res_lsq.x))
f = lambda x: res_lsq.x[0] + res_lsq.x[1] * x

x_p = np.linspace(min(x), max(x), 100)
y_p = f(x_p)

plt.plot(x, y, 'o', label='Дані')
plt.plot(x_p, y_p, 'b', label='МНК (наближення прямою)')
plt.title("МНК (наближення функцією $f(x) = e^{\cos(x)}$)")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import least_squares
import math

def func(x):
    return np.exp(np.cos(x))

x = np.array([i * 0.1 for i in range(1, 11)])
y = np.array([func(xi) for xi in x])
print('x =', x)
print('y =', y)

# Наближення функцією
def fun(a, x, y):
    return a[0] * np.exp(np.cos(a[1] * x)) - y

a0 = np.array([1, 1])
res_lsq = least_squares(fun, x0=a0, args=(x, y))
print("a0 = %.2f, a1 = %.2f" % tuple(res_lsq.x))
f = lambda x: res_lsq.x[0] * np.exp(np.cos(res_lsq.x[1] * x))

x_p = np.linspace(min(x), max(x), 100)
y_p = f(x_p)

plt.plot(x, y, 'o', label='Дані')
plt.plot(x_p, y_p, 'b', label='МНК (наближення функцією $f(x) = e^{\cos(x)}$)')
plt.title("МНК (наближення функцією)")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
